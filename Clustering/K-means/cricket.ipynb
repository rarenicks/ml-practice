```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt

import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
# read the dataset
cricket_df = pd.read_csv("Cricket.csv", sep=",", encoding="ISO-8859-1", header=0)
cricket_df.head()
```

```python
# 2. rescaling
small_df = cricket_df[['SR', 'Ave']]

# instantiate
scaler = StandardScaler()

# fit_transform
small_df_scaled = scaler.fit_transform(small_df)
small_df_scaled.shape
```

```python
# k-means with some arbitrary k
kmeans = KMeans(n_clusters=4, random_state=100)
kmeans.fit(small_df_scaled)
```

```python
# assign the label
cricket_df['cluster_id'] = kmeans.labels_
cricket_df.head()
```
